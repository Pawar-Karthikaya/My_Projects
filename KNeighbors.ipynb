```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Wine dataset
wine = load_wine()
X = wine.data
y = wine.target

# Create a DataFrame for data analysis
wine_df = pd.DataFrame(data=np.c_[wine['data'], wine['target']],
                       columns=list(wine['feature_names']) + ['target'])

# Select a subset of relevant features
selected_features = ['alcohol', 'flavanoids', 'proline']

# Add target to the selected features
selected_features.append('target')

# Create a pairplot for the selected features
sns.pairplot(wine_df[selected_features], hue='target', palette='Set1')
plt.show()

```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a k-NN classifier
k = 3  # Number of neighbors (you can adjust this)
knn_classifier = KNeighborsClassifier(n_neighbors=k)
```

```python
# Fit the model on the training data
knn_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = knn_classifier.predict(X_test)
```

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Generate a classification report and confusion matrix
class_names = wine.target_names
classification_rep = classification_report(y_test, y_pred, target_names=class_names)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Classification Report:\n", classification_rep)
print("Confusion Matrix:\n", conf_matrix)
```
